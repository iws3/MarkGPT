```python
# simple deep learning implementation

import numpy as np
import matplotlib.pyplot as plt
```

```python
# setup data -> 3 inputs -> 2 outputs

N = 100;
print(N)
```

```python
X = np.random.randn(N, 3)
X
```

```python
# Lets make it leaarn y1 = x1 + x2 and y2 = x3*0.5
Y = np.zeros((N, 2))
Y
```

```python
Y[:, 0] = X[:, 0] + X[:, 1]
Y[:, 1] = X[:, 2] * 0.5
print(Y)
```

```python
# Imitilaize the weight
W1=np.random.randn(3,4) * 0.1
print(W1)
```

```python
b1=np.zeros((1, 4))
W2 = np.random.randn(4, 4) * 0.1
b2 = np.zeros((1, 4))
W3 = np.random.randn(4, 2) * 0.1
b3 = np.zeros((1, 2))
```

```python
print(W2)
```

```python
lr=0.01
history=[]
print(history)
print(lr)
```

```python
# 1000 is the epoch, the number of times we want to run
for i in range(1000):
    # forward pass
    z1=X@W1 + b1
    a1=np.maximum(0, z1) # Relu
    z2=a1@W2 + b2 
    a2=np.maximum(0, z2) # Relu
    # print(f"Value of z1: {z1}")
    # print(f"Value of z2: {z2}")
    # print(f"Value of a1: {a1}")
    # print(f"Value of a1: {a1}")

    y_hat=a2@W3 + b3

    # Loss (mse)
    loss=np.mean((y_hat - Y)**2)
    history.append(loss)

    # __Backward pass__
    # layer 3
    dy_hat=2.0*(y_hat-Y)/N
    dW3=a2.T @ dy_hat
    db3=np.sum(dy_hat, axis=0, keepdims=True)
    da2=dy_hat @ W3.T

    # Layer 2
    dz2 = da2.copy()
    dz2[z2 <= 0] = 0 # ReLU backward
    dW2 = a1.T @ dz2
    db2 = np.sum(dz2, axis=0, keepdims=True)
    da1 = dz2 @ W2.T
    
    # Layer 1
    dz1 = da1.copy()
    dz1[z1 <= 0] = 0 # ReLU backward
    dW1 = X.T @ dz1
    db1 = np.sum(dz1, axis=0, keepdims=True)
    # --- UPDATE ---
    W3 -= lr * dW3
    b3 -= lr * db3
    W2 -= lr * dW2
    b2 -= lr * db2
    W1 -= lr * dW1
    b1 -= lr * db1

# 4. VISUALIZATION
plt.figure(figsize=(10, 4))
plt.plot(history)
plt.yscale('log')
plt.title("Loss Training Curve (3-4-4-2 Network)")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.grid(True)
plt.show()
    
```
