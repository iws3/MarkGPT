# Project 1: Sensor Data Analyser (NumPy)
Generate 10,000 simulated sensor readings
(temperature + current + voltage) using
numpy.random. Compute statistics across all
three channels. Apply boolean indexing to
extract anomalies. Save the anomalies to a
separate CSV using only NumPy (savetxt).

Skills: NumPy arrays, vectorisation, boolean
indexing, np.savetxt

```python
import numpy as np

#   DATA SIMULATION 
# We are creating 10,000 fake sensor readings to test our system
n = 10000

# Generating random values of n=10000
current = np.random.uniform(5, 20, n) + np.random.normal(0, 0.2, n)
temp = np.random.normal(25, 5, n) + np.random.normal(0, 1, n)

# We use a math formula to make Voltage drop slightly as Current goes up.
# This mimics how real electricity behaves in a wire
voltage = 230 - 0.5 * current + np.random.normal(0, 1, n)

# Put all three sensors (Temp, Current, Voltage) into one big spreadsheet-style table
data = np.column_stack((temp, current, voltage))

#   STATISTICAL ANALYSIS 
# Find the "normal" behavior: what is the average value and how much does it usually wiggle (spread)?
avg, sd = data.mean(axis=0), data.std(axis=0)

#   ANOMALITY DETECTION
# Look for "weird" data. If a value is very far away from the average (more than 3 steps of "spread"), 
# we flag it as an outlier or a mistake.
anomaly_mask = np.any(np.abs(data - avg) > (3 * sd), axis=1)

# Grab only the rows that were flagged as "weird"
anomalies = data[anomaly_mask]

#   OUTPUT
# Show us the average stats and tell us how many broken/weird readings were found
print(f"Stats (Mean): {avg}")
print(f"Stats (std): {sd}")
print(f"Anomalies found: {len(anomalies)} of {n}")

# Take all those weird readings and save them into a file called "Sensed_Anomalies.csv"
np.savetxt(
    "Sensor_Anomalies.csv", 
    anomalies, 
    delimiter=",", 
    header="Temp (°C),Current (A),Voltage (V)", 
    comments="", 
    fmt="%.2f"
)
print("The filtered anomaly list is saved to a CSV file for reporting.\nCheck the outcome ")
```
