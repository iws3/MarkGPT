```python
import pandas as pd
import numpy as np

# Generate 1,000 rows of data
data = {
    'Temperature_C': np.random.normal(55, 10, 1000),
    'Pressure_PSI': np.random.uniform(90, 120, 1000),
    'Vibration_mm_s': np.random.exponential(0.5, 1000),
    'Humidity_Pct': np.random.normal(45, 5, 1000)
}

df = pd.DataFrame(data)

# Round the entire DataFrame to 2 decimal places
df = df.round(2)

# Save to CSV
df.to_csv("sensor_data_sample.csv", index=False)
print("File created: 'sensor_data_sample.csv' (All values rounded to 2 decimals)")
```
