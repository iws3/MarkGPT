```python
# Week 3 Project 1: Vector Operations Visualiser 
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D


def plot_vectors_demo():
    # Set up a wide figure to hold two side-by-side plots
    fig = plt.figure(figsize=(14, 6))
    
   
    # 1. 2D PLOT: ADDITION & BOUNDARY LINE
  
    ax1 = fig.add_subplot(121)
    
    # Define two sensor vectors
    v1 = np.array([3, 4]) # e.g., Temperature reading
    v2 = np.array([4, 1]) # e.g., Current reading
    v_sum = v1 + v2       # Vector addition: combining the signals[cite: 1]
    
    # Define a 2D Weight Vector (The "Brain's" rule for what is important)[cite: 1]
    w_2d = np.array([1, -1]) 
    
    # Calculate the Decision Boundary line
    # This line is perpendicular to the weight vector
    # Mathematically: y = -(w1/w2) * x
    x_vals = np.linspace(0, 8, 100)
    y_vals = -(w_2d[0] / w_2d[1]) * x_vals
    
    # Draw the arrows starting from (0,0)
    ax1.quiver([0, 0, 0], [0, 0, 0], [v1[0], v2[0], v_sum[0]], [v1[1], v2[1], v_sum[1]], 
               color=['r', 'b', 'g'], scale=1, scale_units='xy', angles='xy', label='Red:V1, Blue:V2, Green:Sum')
    
    # Draw the dashed fence (Decision Boundary)
    ax1.plot(x_vals, y_vals, 'k--', alpha=0.5, label='Decision Boundary')
    
    ax1.set_xlim(0, 8); ax1.set_ylim(0, 8)
    ax1.set_title("2D: Vector Addition & Boundary Line")
    ax1.legend()
    ax1.grid(True, linestyle=':', alpha=0.9)

    
    # 2. 3D PLOT: SPACE & BOUNDARY PLANE
  
    
    ax2 = fig.add_subplot(122, projection='3d')  # '122' means 1 row, 2 columns, this is the 2nd plot
    
  
    v3d = np.array([2, 5, 8])   # 3D Sensor Vector (Temperature, Current, Voltage)
    
    # Define 3D Weight Vector
    
    w_3d = np.array([0.5, 0.2, -0.1]) # This determines the "tilt" of the 3D boundary plane
    
    # Create a grid for the Plane (The "Sheet" that acts as a fence)
    # Equation: w1*x + w2*y + w3*z = 0  => z = -(w1*x + w2*y) / w3
    xx, yy = np.meshgrid(np.linspace(0, 10, 10), np.linspace(0, 10, 10))
    zz = -(w_3d[0] * xx + w_3d[1] * yy) / w_3d[2]
    
    # Plot the 3D sensor arrow
    # The first three (0,0,0) are the start; the next three are the direction
    ax2.quiver(0, 0, 0, v3d[0], v3d[1], v3d[2], color='purple', label='3D Sensor Reading')
    
    # Plot the decision plane (The translucent gray sheet)
    ax2.plot_surface(xx, yy, zz, alpha=0.2, color='black')
    
    # Set the size of the 3D "glass box"
    ax2.set_xlim(0, 10); ax2.set_ylim(0, 10); ax2.set_zlim(0, 10)
    ax2.set_xlabel('Temp'); ax2.set_ylabel('Current'); ax2.set_zlabel('Voltage')
    ax2.set_title("3D: Sensor Space & Decision Plane")
    
    plt.tight_layout()
    plt.show()

# Run the function to see the results
plot_vectors_demo()
```
