# COMPLETE REGRESSION PIPLINE
Every steps from raw data to evaluted model

```python
# polinomial regression
import numpy as np 
from sklearn.model_selection import train_test_split
from sklearn.linear_model import  LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
# - 1. create house price data
np.random.seed(42)
house_size = np.random.uniform(500,3000,100)
price = 200 * house_size + 50000 + np.random.normal(0, 20000, 100)

x= house_size.reshape(-1, 1) # scikit_learnneeds 20 array
y = price
# - 2. split
x_train, x_test, y_train, y_test =train_test_split(
    x, y, test_size = 0.2, random_state=42
)

#- 3. Create -> Train -> Pridict
model = LinearRegression()   # create
model.fit(x_train, y_train)   # Train (finds w0,w1)
y_pred = model.predict(x_test)  # Predict on test set

# - 4. Evaluate
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_pred)

print(f"Slop (w1): {model.coef_[0]:.2f}")  # = 200
print(f"Intercep (w0): {model.intercept_:.2f}") # =50,000
print(f"RMSE:  ${rmse:,.0f}")  # =avg error in $
print(f"R2 SCORE: {r2:.3f}")  # closer to 1 = beter
```
