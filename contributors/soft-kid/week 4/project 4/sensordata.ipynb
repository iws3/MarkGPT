```python
import numpy as np
import pandas as pd

np.random.seed(42)
n = 100
# Features: operational hours, ambient temperature, load percentage, voltage
current = np.random.uniform(30, 5, n) 
ambient = np.random.uniform(20, 45, n) # ambient temp °C
temp = np.random.uniform(25, 50, n)
voltage = np.random.normal(220, 5, n) # supply voltage V
fault = np.random.normal(20, 50, n)
# Convert boolean results (True/False) directly to integers (1/0)
f = (fault > 30).astype(int)

df = pd.DataFrame({
"Current":current,
"Ambient": ambient,
"Temp": temp,
"Voltage": voltage,
"Fault": f
})
# print (f)
df.to_csv("sensor_data.csv")
```
