# Project 2: Energy Audit Dashboard
(Pandas)
Load a week of power meter data from CSV (or
simulate 7 days × 288 readings). Use Pandas
groupby to compute daily consumption totals
(kWh). Identify the peak demand hour of each
day. Print a formatted daily report table.
Skills: Pandas DataFrame, groupby, resample,
describe(), to_csv

```python
import pandas as pd
import numpy as np
from datetime import datetime

# GET REAL TIME
now = datetime.now()

# GENERATING TIMESTAMPS of 7 days haviing 5-minute intervals ending now
time_idx = pd.date_range(end=now, periods=7*288, freq="5min")

# Random kWh readings for these real timestamps
df = pd.DataFrame({"kWh": np.random.uniform(0.5, 5.0, len(time_idx))}, index=time_idx)

# DATA OVERVIEW 
print("--- Data Stats ---")
print(df.describe().round(2))

# SUMMARIZE: Group by Day
daily_totals = df.resample('D').sum()

#  Get the time of highest usage
peak_hours = df.resample('D').apply(lambda x: x.idxmax().strftime('%H:%M'))

# Combine stats, costs, and status
report = daily_totals.copy()
report['Cost ($)'] = (report['kWh'] * 0.15).round(2)
report['Peak Hour'] = peak_hours

weekly_avg = report['kWh'].mean()
report['Status'] = np.where(report['kWh'] > weekly_avg, "High Usage", "Normal")

print(f"Report Generated On: {now.strftime('%Y-%m-%d %H:%M:%S')}")
print(f"Weekly Average: {weekly_avg:.2f} kWh\n")
print(report)
report.to_csv("Energy_Audit.csv")
```
