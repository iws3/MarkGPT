## 7.3 Linear Algebra Operations

```python
import torch
import time

start=time.time();
A=torch.randn(4, 4)
print(A)
print(f"Time it rand was : {time.time() - start:.4f}s")
```

```python
# Determinant
det=torch.linalg.det(A);
# inverse
A_inv=torch.linalg.inv(A)
print(f"The determinant of A is: {det}");
print(f"The inverse of A is: {A_inv}")
```
