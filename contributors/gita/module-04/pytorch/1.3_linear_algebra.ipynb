## 7.3 Linear Algebra Operations

```python
import torch
import time

start=time.time();
A=torch.randn(4, 4)
print(A)
print(f"Time it rand was : {time.time() - start:.4f}s")
```

```python
# Determinant
det=torch.linalg.det(A);
# inverse
A_inv=torch.linalg.inv(A)
print(f"The determinant of A is: {det}");
print(f"The inverse of A is: {A_inv}")
```

```python
eigenvalues, eigenvectors=torch.linalg.eig(A)
print(f"The value of eigen vector is: {eigenvalues}")
```

```python
# Singular Value Decomposition (SVD)
# Used in attention mechanisms, dimensionality reduction
U, S, Vh = torch.linalg.svd(A)
U
```

```python
torch.linalg.norm(A) # Frobenius norm of matrix
```

```python
torch.linalg.norm(A, dim=1)  # l2 norm of each row
```

```python
# solve linear system : Ax=b
b=torch.randn(4)
x=torch.linalg.solve(A, b)
x
```
