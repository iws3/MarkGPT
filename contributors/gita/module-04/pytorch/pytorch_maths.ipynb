## 7.1 Element-wise Operations


```python
import torch

a=torch.tensor([1.0,4.0, 9.0, 16.0])
b=torch.tensor([2.0,2.0, 3.0, 4.0])
print(a)
print(b)
```

```python
# Arithmeic

print(a+b)
print(a-b)
print(a*b)
print(a/b)
print(a**b)
# floor division
print(a//b)
# mudolo
print(a%b)
```

```python
# Maths functions
print(torch.sqrt(a))
print(torch.exp(a))  # e^a element-wise
print(torch.log(a)) # natural log
print(torch.log2(a)) # log base 2
print(torch.abs(a))
print(torch.ceil(a)) #round up
print(torch.floor(a)) #round down
print(torch.clip(a, 2, 8)) # clip a between [2, 8]
```

## 7.2 Broadcasting

Broadcasting is how PyTorch (and NumPy) handle math between tensors of different shapes. It's essential for
understanding how batches and bias terms work.


```python
# Rule: dimensions are aligned from RIGHT, and dimensions of size 1 stretch

a=torch.ones(3,4)
b=torch.ones(4)
print(a)
print(b)
print(a+b) # [3,4] + [4] -> b broadcast to [3, 4]
```

```python
# Concrete example: adding bias to a batch

batch=torch.randn(32, 10); # 32 samples, 10 features
bias=torch.randn(10)  # One bias per feature
result=batch  + bias;
print(f"Result is: {result}")
print(batch)
```

```python
# Another example: normalizing each sample
data=torch.randn(32, 100)     # [32, 100]
print(data)
# print(data.size())
means=data.mean(dim=1, keepdim=True) # [32, 1]  -> keepdim is crucial here
stds=data.std(dim=1, keepdim=True) # [32, 1]
normalized=(data-means)
```
