## 7.4 Activation Functions (Mathematical Forms)
These are applied element-wise and introduce non-linearity — without them, stacking linear layers would just
be one linear layer.

import torch 
import torch.nn.functional as nn

x=torch.randn(5)
print(x)

```python
# ReLU: f(x)=max(0, x)
# Most common activation , simple fast, Problem: "dying ReLU" for negative inputs.

relu_out=nn.relu(x)
relu_out=torch.clamp(x, min=0)  # Same thing manually

print(relu_out)
```

```python
# Sigmoid: f(x)=1/(1-e^(-x))

# output range is (0, 1), Used for binary classification outputs.
# Problem: Vanishing gradients in deep networks
sigmoid_out=torch.sigmoid(x)
print(sigmoid_out);
```
