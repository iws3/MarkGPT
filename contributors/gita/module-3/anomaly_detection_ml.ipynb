```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
# GENERATE SYNTHETIC DATA - FIRST : enable reproducity
np.random.seed(42)
# In practice we will load the data frame here
def generate_dataset(n_normal=300, n_anomaly=20, seed=42):
    rng=np.random.RandomState(seed)
    cluster_a=np.random.multivariate_normal(
        mean=[0.0, 0.0],
        cov=[[1.2, 0.5], [0.5, 0.9]],
        size=int(n_normal * 0.55)
    )
    print(f"printing cluster_a: {cluster_a}")
    return cluster_a
```

```python
cluster_a=generate_dataset()
def visualize_dataset(cluster_a=cluster_a):
    x1=cluster_a[:, 0]
    y1=cluster_a[:, 1]
    # create subplot
    fig, axes=plt.subplots(1, 2, figsize=(12, 5))
    # Plot 1: Actual cluster
    axes[0].scatter(x1, y1, alpha=0.8)
visualize_dataset()
    
```
